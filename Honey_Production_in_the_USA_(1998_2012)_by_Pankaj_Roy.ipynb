Use the Honey Production in the USA Dataset and solve the following question by using the dataset,to download the dataset click on the dataset name.Topic:To visualize howhoney production is changed over the years(1998-2021)in the UnitedStates.

Background:In 2006,global concern was raised over the rapid decline in the honeybee population,an integral component of American honey agriculture.Large numbers of hives were lost to Colony Collapse Disorder,a phenomenon of disappearing worker bees causing the remaining hive colony to collapse.Speculation as to the cause of this disorder points to hive diseases and pesticides harming the pollinators,though no overall consensus has been reached.The U.S.used to locally produce over half the honey it consumes per year.Now,honey mostly comes from overseas,with 350 of the 400 million pounds of honey consumed every year originating from imports.This
dataset provides insight into honeyproduction supply and demand in America from 1998 to 2021.

# 1] How has honey production yield changed from 1998 to 2021?

```python
x=df.groupby(df['year'])['yieldpercol'].mean()
x
```

**Ploting line chart using matplotlib**

```python
plt.figure(figsize=(8,6))
plt.plot(x.index,x.values,color='blue',marker='d')
plt.title('Year vs Average honey production yield',fontsize=15,color='red')
plt.xlabel('Year',fontsize=15)
plt.ylabel('Average honey production yield',fontsize=15)
plt.grid()
plt.show()
```

**Ploting line chart using Seaborn**

```python
plt.figure(figsize=(8,6))
sns.pointplot(x=x.index,y=x.values)
sns.set_style('darkgrid')
sns.set_context('notebook')
plt.xticks(rotation =45);
plt.title('Year vs Average honey production yield',fontsize=15,color='red')
plt.xlabel('Year',fontsize=15)
plt.ylabel('Average honey production yield',fontsize=15)
plt.grid()
plt.show()
```

# 3] Does the data show any trends in terms of the number of honey producing colonies and yield per colony before 2006,which was when concern over Colony Collapse Disorder spread nationwide?

```python
d=df[df['year']<2006]
d1=d[['numcol','yieldpercol']]
d1
```

```python
plt.figure(figsize=(8,6))
plt.scatter(d1['numcol'],d1['yieldpercol'])
plt.title('numcol VS yield per colony',fontsize=15,color='blue')
plt.xlabel('Number of honey producing colonies')
plt.ylabel('yield per colony')
plt.grid()
plt.legend()
plt.show()
```

# 4] Are there any patterns that can be observed between total honey production and value of production every year?

```python
x=df.groupby(['year']).agg({'prodvalue':['sum'],'totalprod':['sum']})
x.columns
```

**Ploting line chart using Seaborn**

```python
plt.figure(figsize=(8,6))
sns.pointplot(x=x.index,y=x[('prodvalue', 'sum')],color='blue',label='Total Production Value')
sns.pointplot(x=x.index,y=x[('totalprod', 'sum')],color='orange',label='Total Production')
plt.title('Year vs Total Production and Value of Production',fontsize=15,color='green')
plt.xlabel('Year',fontsize=15)
plt.ylabel('Total Production and Value of Production',fontsize=15)
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

**Top 10 Cities which have maximum Honey Stocks present overall years**

Conclusion-Maximum quantity of honey stocks held by producers in South Dekota

```python
r=df.groupby(df['State'])['stocks'].sum()
r1=r.sort_values(ascending=False)
r2=r1.head(10)
sns.set_palette('bright')
sns.set_style('whitegrid')
sns.set_context('paper')
plt.figure(figsize=(8,6))
sns.barplot(x=r2.values,y=r2.index)
plt.xticks(rotation=90)
plt.title('State vs Total Stocks',fontsize=15,color='red')
plt.xlabel('State',fontsize=15)
plt.ylabel('Total Stocks',fontsize=15)
plt.show()
```

**Total honey stocks trends in each year**

Conclusion-Maximum Quantity honey available in Year 2000 

```python
x=df.groupby(df['year'])['stocks'].sum()
plt.figure(figsize=(8,6))
sns.set_style('darkgrid')
sns.set_context('notebook')
sns.set_palette('magma')
sns.barplot(x=x.index,y=x.values)
plt.xticks(rotation=90)
plt.title('Year vs Total Stocks',fontsize=15,color='red')
plt.xlabel('Year',fontsize=15)
plt.ylabel('Total Stocks',fontsize=15)
plt.show()
```
